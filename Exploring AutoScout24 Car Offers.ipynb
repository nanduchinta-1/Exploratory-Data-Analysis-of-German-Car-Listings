## Module 1
### Task 1: Reading and Analyzing the Data.
Before we start our analysis, first, we need to view the dataset. It is essential to view the data and check the columns. Let's take a look.

```python
import warnings
warnings.filterwarnings("ignore")

#--- Import Pandas ---
import pandas as pd
#--- Read in dataset ----
df = pd.read_csv("./car_data.csv")

# ---WRITE YOUR CODE FOR TASK 1 ---
#--- Inspect data ---
df
```

### Task 2: Dropping Column from DataFrame.
Great!!! We have our dataset loaded. Now, we need to remove the unwanted columns. So, let's remove those columns.

```python
# Dropping the first column
#--- WRITE YOUR CODE FOR TASK 2 ---
df = df.iloc[:,1:]

#--- Inspect data ---
df
```

### Task 3: Counting Null Values in the Data.
Wow!!! We have successfully remove the column. Now, we need to ensure that the dataset does not contain any null values. So, let's check for null values in our dataset.

```python
#--- WRITE YOUR CODE FOR TASK 3 ---
null_values = df.isnull().sum()

#--- Inspect data ---
null_values
```

### Task 4:  Analyzing Missing Fuel Consumption Data by Fuel Type.
Incredible!!! We've discovered null values in our dataset. Now, our next step is to ensure the dataset is free of any missing values. To do this, we're checking for null values specifically in the 'fuel_consumption_l_100km' column. Let's see how many missing values we have for each fuel type in our dataset.

```python
#--- WRITE YOUR CODE FOR TASK 4 ---
filtered_df = df[df.fuel_consumption_l_100km.isnull()]

# Group filtered rows by 'fuel_type' and count occurrences of each 'fuel_type'
missing_df = filtered_df['fuel_type'].value_counts()

# Create a DataFrame with counts of missing values indexed by 'fuel_type'

# Display or use the 'missing_df' DataFrame as needed
missing_df
```

### Task 5:  Handling Missing Fuel Consumption Data.
Fantastic!!! We've found null values in fuel_type. Now, we need to fill in these missing values to ensure our dataset is ready for further analysis. Let's proceed with filling them in.

```python
#--- WRITE YOUR CODE FOR TASK 5 ---
for index, row in df.iterrows():
    # Check if 'fuel_consumption_l_100km' is null and 'fuel_type' is 'Electric', 'Hydrogen', or 'Ethanol'
    if pd.isnull(row['fuel_consumption_l_100km']) and row['fuel_type'] in ['Electric', 'Hydrogen', 'Ethanol']:
        # Replace the missing value with float(0)
        df.at[index, 'fuel_consumption_l_100km'] = float(0)
#--- Inspect data ---
df
```

### Task 6: Removing Unknown and Other Fuel Types.
Wow!!! We've successfully filled the values. Now, we need to remove the unwanted rows. By filtering out these entries, we streamline our dataset for clearer analysis. Let's take a look at the refined dataset without these entries.

```python
#--- WRITE YOUR CODE FOR TASK 6 ---
df = df[df["fuel_type"]!="Unknown"]
df = df[df["fuel_type"]!="Other"]
#--- Inspect data ---
df
```

### Task 7: Removing Rows with Missing Values.
Great!!! We've successfully removed the unwanted values. Now, we need to drop the rows containing any missing values. This helps maintain data quality and prepares our dataset for thorough analysis. Let's proceed with the cleaned dataset.

```python
#--- WRITE YOUR CODE FOR TASK 7 ---
df.dropna(inplace=True)

#--- Inspect data ---
df
```

## Module 2
### Task 1:  Analyzing Car Listings by Year.
Incredible!!! We've successfully removed the unwanted rows. Now, we need to examine the distribution of car listings by year.Let's explore it.

```python
#--- WRITE YOUR CODE FOR TASK 1 ---
year_values = df.year.value_counts()

#--- Inspect data ---
year_values
```

### Task 2: Filtering Years in Dataset.
Fantastic!!! We've analyzed the years. Now, we need to create a list of recent years. By selecting only the year values, we aim for better data for analysis. Let's get to work on it.

```python
#--- WRITE YOUR CODE FOR TASK 2 ---
yearsToFilter = list(year_values.index.unique().sort_values()[1:30])

#--- Inspect data ---
yearsToFilter
```

### Task 3: Filtering Dataset by Years.
Wow!!! We've successfully created a list of years. Now, we need to filtering our dataset to include only entries from year values. Applying this filter helps us focus our analysis on more cleaned data. Let's filter the dataset accordingly.

```python
filt = []

#--- WRITE YOUR CODE FOR TASK 3 ---
for index, row in df.iterrows():
    # Check if the year value of each row is in the list yearsToFilter
    filt.append(row['year'] in yearsToFilter)
#--- Inspect data ---
df = df[filt]
df
```

### Task 4: Extracting Fuel Consumption Range for Electric Cars.
Great!!! We've successfully filtered the years. Now, we need to process the fuel_consumption_l_100km column. By doing this it helps us focus our analysis on more cleaned data. Let's process the dataset accordingly.

```python
# Extract Range from 'fuel_consumption_g_km' for Electric Cars
#--- WRITE YOUR CODE FOR TASK 4 ---
# df = ...
df['fuel_consumption_l_100km'] = df.apply(lambda row: float(row['fuel_consumption_g_km'][:3]) if 'Reichweite' in row['fuel_consumption_g_km'] and row['fuel_type'] == 'Electric' and 'k' not in row['fuel_consumption_g_km'][:3] else row['fuel_consumption_l_100km'], axis=1)

#--- Inspect data ---
df
```

### Task 5: Standardizing Consumption Value Formatting.
Fantastic!!! We've successfully processed the data. Now, we need to process the fuel_consumption_l_100km column again for formatting. By doing this it helps us focus our analysis on more cleaned data. Let's process the dataset accordingly.

```python
# Handle Formatting of Consumption Values
#--- WRITE YOUR CODE FOR TASK 5 ---
df['fuel_consumption_l_100km'] = df.apply(lambda row: 
                                           float(row['fuel_consumption_l_100km'].replace(' kWh/100 km', '').replace(',', '.').split(' ')[0])
                                           if isinstance(row['fuel_consumption_l_100km'], str) and ('kWh/100 km' in row['fuel_consumption_l_100km']) 
                                           else row['fuel_consumption_l_100km'], axis=1)
#--- Inspect data ---
df
```



### Task 6: Standardizing Consumption Value Formatting.
Incredible!!! We've successfully formatted the data. Now, we need to identify values in the 'fuel_consumption_l_100km' column that need to be dropped. By doing this step helps us identify and handle irrelevant data in our dataset. Let's mark these values for removal in the dataset.

```python
#--- WRITE YOUR CODE FOR TASK 6 ---
df['fuel_consumption_l_100km'] = df.apply(lambda row: 
                                          'To Drop' if (not isinstance(row['fuel_consumption_l_100km'], float) or row['fuel_consumption_l_100km'] == 0) and row['fuel_type'] == 'Electric' 
                                          else row['fuel_consumption_l_100km'], axis=1)

#--- Inspect data ---
df
```

### Task 7: Removing Rows Marked for Dropping.
Wow!!! We've successfully processed the data. Now, we need to remove rows from the dataset that were marked in the last task. By filtering out these rows, we ensure the integrity of our dataset for further analysis. Let's proceed with removing these rows from the dataset.

```python
# Drop Rows Marked for Dropping
#--- WRITE YOUR CODE FOR TASK 7 ---
df = df[df['fuel_consumption_l_100km']!='To Drop'] 

#--- Inspect data ---
df
```

## Module 3

### Task 1: Standardizing Consumption Value Formatting for Liters per 100 km.
Great!!! We've successfully dropped the unwanted entries. Now, we need to remove the ' l/100 km' suffix and replace any commas with periods to ensure consistent formatting. This step enhances the clarity and accuracy of our dataset for analysis. Let's implement this formatting adjustment to the dataset.

```python
# Handle Formatting of Consumption Values for liters per 100 km
#--- WRITE YOUR CODE FOR TASK 1 ---
df['fuel_consumption_l_100km'] = df.apply(lambda row: 
                                           float(row['fuel_consumption_l_100km'].replace(' l/100 km', '').replace(',', '.').split(' ')[0])
                                           if isinstance(row['fuel_consumption_l_100km'], str) and ('l/100 km' in row['fuel_consumption_l_100km']) and ('-' not in row['fuel_consumption_l_100km'])
                                           else str(row['fuel_consumption_l_100km'])
                                           if isinstance(row['fuel_consumption_l_100km'], str) and ('l/100 km' not in row['fuel_consumption_l_100km']) and (not row['fuel_consumption_l_100km'].isdigit()) and ('-' in row['fuel_consumption_l_100km'])
                                           else row['fuel_consumption_l_100km'] , axis=1)

#--- Inspect data ---
df
```

### Task 2: Standardizing Consumption Value Formatting for Kilograms per 100 km.
Incredible!!! We've successfully dropped the suffix. Now, we need to remove the ' kg/100 km' suffix and replace any commas with periods to ensure consistent formatting. This step enhances the clarity and accuracy of our dataset for analysis. Let's implement this formatting adjustment to the dataset.

```python
# Handle Formatting of Consumption Values for kilograms per 100 km
#--- WRITE YOUR CODE FOR TASK 2 ---
df["fuel_consumption_l_100km"] = df.apply(lambda row: 
                float(row["fuel_consumption_l_100km"].replace(" kg/100 km","").replace(",",".").split(" ")[0])
                if isinstance(row["fuel_consumption_l_100km"],str) and ("kg/100 km" in row["fuel_consumption_l_100km"]) and ("-" not in row["fuel_consumption_l_100km"])
                else str(row['fuel_consumption_l_100km'])
                if isinstance(row['fuel_consumption_l_100km'], str) and ('kg/100 km' not in row['fuel_consumption_l_100km']) and (not row['fuel_consumption_l_100km'].isdigit()) and ('-' in row['fuel_consumption_l_100km'])
                else row['fuel_consumption_l_100km'] , axis=1)

#--- Inspect data ---
df
```

### Task 3: Filtering DataFrame for Float Values.
Fantastic!!! We've successfully dropped the suffix again. Now, we need to retain rows where the 'fuel_consumption_l_100km' column contains float values. By doing this, we ensure data integrity and prepare the dataset for further analysis. Let's take a look.

```python

# Filter DataFrame to Retain Rows with Float Values in 'fuel_consumption_l_100km'
#--- WRITE YOUR CODE FOR TASK 3 ---
fl = []
for i, row in df.iterrows():
    if (not isinstance(row['fuel_consumption_l_100km'], float)):
        fl.append(i)
df = df.drop(index=fl)
df
```

### Task 4: Converting Data Types.
Wow!!! We've successfully filtered the DataFrame. Now, we need to convert the data types of numbers to integer and floating point. By doing this, we ensure data integrity and prepare the dataset for further analysis. Let's apply these data type conversions to the DataFrame.

```python
#--- WRITE YOUR CODE FOR TASK 4 ---
df = df.astype({'price_in_euro': 'int',
          'power_kw':'int',
          'fuel_consumption_l_100km':'float'})

#--- Inspect data ---
df.info()
```

## Module 4
### Task 1: Visualizing Number of Car Listings by Brand.
Incredible!!! We've successfully cleaned the dataframe for the analysis. Now, we're visualizing the distribution of car listings across different brands. By creating a horizontal bar plot, we showcase the number of listings for each brand in our dataset. This visualization provides insights into the popularity and presence of various car brands in the market. Let's explore the distribution of car listings per brand.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

brand_counts = df.brand.value_counts()
#--- WRITE YOUR CODE FOR TASK 1 ---
plot1 = plt.barh(brand_counts.index,brand_counts.values)
plt.xlabel('Number of Listings')
plt.ylabel('Brand')
plt.title('Number of Listings Per Brand')

#--- Inspect data ---
plt.show()
```

### Task 2: Analyzing Average Price Per Brand.
Great!!! We've successfully plotted the distribution of car listings across different brands. Now, we're analyzing the average price of car listings for each brand in our dataset. This visualization helps us understand the relative pricing of cars from various manufacturers. Let's explore the average price per brand in our dataset.

```python
plt.figure(figsize=(10, 6))
brandPrice = df.groupby(['brand'])[['price_in_euro']].mean()
#--- WRITE YOUR CODE FOR TASK 2 ---
plot2 = plt.barh(brand_counts.index,[brandPrice.loc[i][0] for i in brand_counts.index])
plt.xlabel("Average Price (EUR)")
plt.title('Average Price Per Brand')
plt.show()
#--- Inspect data ---
```

### Task 3: Analyzing Efficiency of Internal Combustion Engine (ICE) Cars Per Brand.
Fantastic!!! We've analyzed the average price of car listings for each brand. Now, we're analyzing the efficiency of internal combustion engine (ICE) cars for selected fuel types across different car brands. This visualization helps us understand the relative fuel efficiency of ICE cars across different brands in our dataset. Let's explore the efficiency of ICE cars per brand.

```python
plt.figure(figsize=(10, 6))
ice_df = df[df['fuel_type'].isin(['Petrol', 'Diesel', 'Hybrid', 'Diesel Hybrid'])]
brandEff = ice_df.groupby('brand')[['fuel_consumption_l_100km']].mean()
#--- WRITE YOUR CODE FOR TASK 3 ---
plot3 = plt.barh(brand_counts.index,[brandEff.loc[i][0] for i in brand_counts.index])
plt.xlabel('Litres of Fuel per 100km')
plt.ylabel('Brand')
plt.title('Efficiency of ICEs Per Brand')

#--- Inspect data ---
plt.show()
```

### Task 4: Analyzing Average Mileage Per Brand.
Wow!!! We've analyzed the efficiency of internal combustion engine (ICE) cars. Now, we're analyzing the average mileage of car listings for each brand in our dataset. This visualization helps us understand the relative mileage performance of cars from various manufacturers. Let's explore the average mileage per brand in our dataset.

```python
plt.figure(figsize=(10, 6))
brandMiles = df.groupby('brand')[['mileage_in_km']].mean()

#--- WRITE YOUR CODE FOR TASK 4 ---
plot4 = plt.barh(brand_counts.index,[brandMiles.loc[i][0] for i in brand_counts.index])
plt.xlabel('Mileage (km)')
plt.title('Average Mileage Per Brand')
#--- Inspect data ---
plt.show()
```
